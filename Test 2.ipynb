```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

#A sample differential equation "dy/dt = -2yt"
def dydt(y,t):
    dydt=-2*y*t
    return dydt

start=0
end=2
t=np.linspace(start,end,10)

y0=1

y = sc.integrate.odeint(dydt,y0,t)

plt.figure()
plt.plot(t,y,'r',label='dy/dt')
plt.title('y(t) versus (t)')
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
import requests

def hourly_temperature_data(latitude, longitude, timezone, forecast_days):
    api_url = f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&hourly=temperature_2m&timezone={timezone}&forecast_days={forecast_days}"
    response = requests.get(api_url)

    if response.ok:
        try:
            hourly_temperature_data = response.json()
            
            while True:
                number_of_day = input("Enter the number of day:")
                if not number_of_day:
                    break
                    
                temperature = hourly_temperature_data.get(number_of_day, {}).get("temperature_2m")
                
                if temperature is not None:
                    print(f"Temperature of {number_of_day}: {temperature}")
                else:
                    print(f"{number_of_day} is not found in the data.")
        except ValueError:
            print("Error: Unable to parse JSON response.")
    else:
        print(f"Error: {response.status_code} - {response.reason}")

hourly_temperature_data(latitude=2.5, longitude=112.5, timezone="Asia%2FSingapore", forecast_days=1)

import math
from matplotlib import pyplot as plt

def plot_hourly_temperature_data(hourly_temperature_data):
    x = list(range(len(hourly_temperature_data)))
    y = [entry.get("temperature_2m", math.nan) for entry in hourly_temperature_data.values()]

    plt.plot(x, y, color='red')
    plt.title('Hourly Temperature Data')
    plt.xlabel('Time (hours)')
    plt.ylabel('Temperature (Â°C)')
    plt.grid(True)
    plt.show()

plot_hourly_temperature_data(hourly_temperature_data)
```
